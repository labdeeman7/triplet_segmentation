```python
import os
import json
import shutil

def update_labelme_with_triplets(labelme_dir, triplet_dict, output_dir):
    os.makedirs(output_dir, exist_ok=True)
    shutil.copytree(labelme_dir, output_dir, dirs_exist_ok=True)  

    for key, triplet in triplet_dict.items():
        frame_name, instrument, group_id_str, _, _ = key.split(",")
        group_id = int(group_id_str)
        verb, target = triplet.split(",")[1:]  # Extract verb and target
        
        assert triplet.split(",")[0] == instrument, f'there is a problem we have different instruments {triplet.split(",")[0]} and instrument'  

        json_path = os.path.join(output_dir, f"{frame_name}.json")
        if not os.path.exists(json_path):
            print(f"⚠️ Missing JSON for {frame_name}")
            continue

        with open(json_path, "r") as f:
            data = json.load(f)

        modified = False
        for shape in data.get("shapes", []):
            if shape["label"] == instrument and shape.get("group_id") == group_id:
                # Add fields
                shape["instrument"] = instrument
                shape["verb"] = verb
                shape["target"] = target
                shape["label"] = triplet  # set full triplet as label
                modified = True

        if modified:
            out_path = os.path.join(output_dir, f"{frame_name}.json")
            with open(out_path, "w") as f:
                json.dump(data, f, indent=2)
            print(f"✅ Updated {frame_name}")
        else:
            print(f"⚠️ No matching shape found for {key} in {frame_name}")

```

```python
labelme_dir = '../../datasets/cholecinstanceseg_extension/matching_issue_tripletsegmentation_extension/ann_dir'
triplet_dict_path = 'outputs/matched_tripletsegmentation_extension.json'
output_dir = '../../datasets/cholecinstanceseg_extension/added_the_triplet_infromation_for_images_with_matching_for_extension'

triplet_dict = read_from_json(triplet_dict_path)

update_labelme_with_triplets(
    labelme_dir=labelme_dir,
    triplet_dict=triplet_dict,
    output_dir=output_dir
)
```

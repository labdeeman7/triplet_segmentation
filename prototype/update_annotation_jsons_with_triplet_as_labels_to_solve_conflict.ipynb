```python
import os
import json
import shutil

def update_labelme_with_triplets_for_conflict(labelme_dir, conflict_matched_triplet_dict, output_dir):
    os.makedirs(output_dir, exist_ok=True)
    shutil.copytree(labelme_dir, output_dir, dirs_exist_ok=True)  

    for key, triplet in conflict_matched_triplet_dict.items():
        frame_name, instrument, group_id_str = key.split(",")
        group_id = int(group_id_str)
        verb, target = triplet.split(",")[1:]  # Extract verb and target
        
        assert triplet.split(",")[0] == instrument, f'there is a problem we have different instruments {triplet.split(",")[0]} and instrument'  

        json_path = os.path.join(output_dir, f"{frame_name}.json")
        if not os.path.exists(json_path):
            print(f"⚠️ Missing JSON for {json_path}")
            continue

        with open(json_path, "r") as f:
            data = json.load(f)

        modified = False
        for shape in data.get("shapes", []):                
            if shape["label"] == instrument and shape.get("group_id") == group_id:
                # Add fields
                shape["instrument"] = instrument
                shape["verb"] = verb
                shape["target"] = target
                shape["label"] = triplet  # set full triplet as label
                modified = True     
        
        if group_id == 1000: 
            extra_shape = {}
            extra_shape["group_id"] = group_id
            extra_shape["description"] = ""
            extra_shape["shape_type"] = "polygon"
            extra_shape["flags"] = None
            extra_shape["points"] = [[300, 300], [500, 300], [500, 500],[300, 500]]
            extra_shape["instrument"] = instrument
            extra_shape["verb"] = verb
            extra_shape["target"] = target
            extra_shape["label"] = triplet  # set full triplet as label
            
            data["shapes"].append(extra_shape)  
            modified = True        

        if modified:
            out_path = os.path.join(output_dir, f"{frame_name}.json")
            with open(out_path, "w") as f:
                json.dump(data, f, indent=2)
            print(f"✅ Updated {frame_name}")
        else:
            raise ValueError(f"⚠️ No updates found for  {key} in json_path")

```

```python
labelme_dir = '../../datasets/matching_and_conflict_for_current_cholecinstanceseg/2_disagreement_issues/ann_dir'
conflict_matched_triplet_dict_path = 'outputs/conflict_matching_tripletsegmentation_v1.json'
output_dir = '../../datasets/matching_and_conflict_for_current_cholecinstanceseg/added_the_triplet_infromation_for_conflict/ann_dir'

conflict_matched_triplet_dict = read_from_json(conflict_matched_triplet_dict_path)

update_labelme_with_triplets_for_conflict(
    labelme_dir=labelme_dir,
    conflict_matched_triplet_dict=conflict_matched_triplet_dict,
    output_dir=output_dir
)
```

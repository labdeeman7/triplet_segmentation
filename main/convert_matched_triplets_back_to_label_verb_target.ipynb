```python
from utils.general.read_files import read_from_json
from utils.general.save_files import save_to_json
from utils.general.dataset_variables import TripletSegmentationVariables

TRIPLETS = list(TripletSegmentationVariables.categories['triplet'].values())
```

```python
import os
import json
import shutil
from os.path import join
```

```python
def format_labelme_with_triplets(matched_ann_dir, matched_img_dir, image_formated_correctly_dir, ann_formated_correctly_dir):
    os.makedirs(ann_formated_correctly_dir, exist_ok=True)
    os.makedirs(image_formated_correctly_dir, exist_ok=True)
    
    ann_files = os.listdir(matched_ann_dir)
    
    
    
    for ann_file in ann_files:
        ann_path = join(matched_ann_dir, ann_file)   
        save_ann_path =  join(ann_formated_correctly_dir, ann_file)  
        
        img_name = ann_file.replace('json', 'png')
        img_path = join(matched_img_dir, img_name) 
        
             
        data = read_from_json(ann_path)
        
        is_snare = False        
        for shape in data.get("shapes", []):   
            triplet=shape["label"]   
            
            if triplet == 'snare,null_verb,null_target':
                is_snare = True  
                break
            assert triplet in TRIPLETS, f'{triplet} for {ann_file} does not exist' 
            instrument, verb, target = triplet.split(",")
            
            
            shape["label"] = instrument
            shape["verb"] = verb
            shape["target"] = target
    
        
        if not is_snare:
            save_to_json(data=data, json_file_path=save_ann_path)
            shutil.copy(img_path, image_formated_correctly_dir)
            print(f"Saved {ann_file}")

```

```python
matched_ann_dir = '../../datasets/matching_and_conflict_for_current_cholecinstanceseg/solved_matching_and_conflict/ann_dir'
matched_img_dir = '../../datasets/matching_and_conflict_for_current_cholecinstanceseg/solved_matching_and_conflict/img_dir'


ann_formated_correctly_dir = '../../datasets/matching_and_conflict_for_current_cholecinstanceseg/solved_matching_and_conflict/ann_formated_correctly'
image_formated_correctly_dir = '../../datasets/matching_and_conflict_for_current_cholecinstanceseg/solved_matching_and_conflict/img_formated_correctly'


format_labelme_with_triplets(matched_ann_dir, matched_img_dir, image_formated_correctly_dir, ann_formated_correctly_dir)
```
